```python
import pandas as pd
```

```python
nat_baby_names = pd.read_csv('NationalNames.csv')
state_baby_names = pd.read_csv('StateNames.csv')
race_names = pd.read_csv('wiki_name_race.csv')
```

```python
nat_baby_names['Name'] = nat_baby_names['Name'].str.lower()
nat_baby_names.head(5)
```

```python
state_baby_names['Name'] = state_baby_names['Name'].str.lower()
state_baby_names.head(5)
```

```python
race_names = race_names[['name_first', 'race']]
race_names.head(5)
```

```python
name_race_counts = race_names.groupby(['name_first', 'race']).size().reset_index(name='counts_name_race')

name_totals_df = name_race_counts.groupby(['name_first'])['counts_name_race'].sum().to_frame().reset_index()
name_totals_df
name_totals_df['counts_name'] = name_totals_df['counts_name_race']
name_totals_df = name_totals_df[['name_first', 'counts_name']]

counts_name_race = pd.merge(name_race_counts, name_totals_df, on=['name_first', 'name_first'])
counts_name_race.head()
```

```python
counts_name_race['racial_proportion'] = counts_name_race['counts_name_race'] / counts_name_race['counts_name']
counts_name_race = counts_name_race.sort_values(by=['counts_name'], ascending=False)
counts_name_race = counts_name_race.reset_index().drop('index', axis=1)
counts_name_race.head(20)
```

```python
counts_name_race.to_csv('first_name_racial_proportions.csv')
```
